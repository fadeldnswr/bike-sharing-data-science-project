```python
Q1 = dataFrame.quantile(0.25)
Q3 = dataFrame.quantile(0.75)
IQR = Q3 - Q1

# Upper Bound 
upperBound = Q3 + 1.5*IQR
lowerBound = Q1 - 1.5*IQR

lowerCondition = dataFrame < lowerBound
upperCondition = dataFrame > upperBound

dataFrame.drop(dataFrame[upperCondition].index, inplace=True)
dataFrame.drop(dataFrame[lowerCondition].index, inplace=True)
```
