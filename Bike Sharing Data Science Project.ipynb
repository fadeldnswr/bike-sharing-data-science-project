## Data Wrangling

### Gathering Data
- Proses gathering data disini terdapat proses merging antara dua dataset, yaitu day.csv dan hour.csv. Hasil gabungan kedua dataset ini akan menghasilkan dataset baru, yaitu bike_sharing_df dengan kolom-kolom yang sama, namun terdapat kolom "hr" yang hanya ada di dataset hour.csv.

- Disini saya juga melakukan proses re-arrangement dataset yang berfungsi untuk mengubah urutan kolom "hr" yang tadinya berada pada posisi paling belakang menjadi di sebelah kanan kolom "dteday".

```python
# Gathering two dataset
bike_per_day_df = pd.read_csv("day.csv")
bike_per_hour_df = pd.read_csv("hour.csv")\

# print(f"Dataset Bike per Day\n {bike_per_day_df.head()}\n")
# print(f"Dataset Bike per Hour\n {bike_per_hour_df.head()}")

# Merging two dataset
bike_sharing_df = pd.merge(
    left=bike_per_day_df,
    right=bike_per_hour_df[["dteday","hr"]],
    how="left",
    on="dteday"
)

# Re-arrange Dataset
columns = bike_sharing_df.columns.tolist()
columns.insert(2, columns.pop(columns.index("hr")))
bike_sharing_df = bike_sharing_df[columns]

print(f"Bike Sharing Dataset \n {bike_sharing_df.head()}")
```

**Insight:**
- Dataset bike_per_day_df dan bike_per_hour_df memiliki kolom yang sama, yang membedakan hanyalah kolom hour pada bike_per_hour_df.
- Melakukan proses merging pada kedua dataset tersebut, sehingga hanya akan ada satu dataset yaitu bike_sharing_df.
- Melakukan proses re-arrange dari dataset gabungan yang baru. Hal ini dilakukan karena kolom "hr" berada di paling belakang. Kolom "hr" dipindahkan ke samping kanan "dteday"

### Assessing Data

- Mengecek dataset bike_sharing_df jika terdapat nilai yang duplikat atau nilai null.

```python
# Checking Null Values
print(f"Null Values from Dataset \n{bike_sharing_df.isnull().sum()}\n")

# Checking Duplicated Values
print(f"Duplicated Values from Dataset : {bike_sharing_df.duplicated().sum()}\n")
```

### Cleaning Data

- Mengecek sekaligus memperbaiki jika terdapat format date and time yang salah.

```python
# Fix the Date and Time Format
bike_sharing_df["dteday"] = pd.to_datetime(bike_sharing_df["dteday"], format="%Y-%m-%d")
incorrectDate = bike_sharing_df[bike_sharing_df["dteday"].isna()]
print(incorrectDate)
```

- Membuat batas atas dan bawah yang diperlukan untuk menghapus nilai outlier yang berada pada kondisi tersebut.

```python
# Drop Outlier for casual columns
Q1 = bike_sharing_df["casual"].quantile(0.25)
Q3 = bike_sharing_df["casual"].quantile(0.75)

IQR = Q3 - Q1
upperBound = Q3 + 1.5 * IQR
lowerBound = Q1 - 1.5 * IQR

upperCondition = bike_sharing_df["casual"] > upperBound
lowerCondition = bike_sharing_df["casual"] < lowerBound
```

- Trimming dataset yang telah di cek nilai outliernya.

```python
# Trimming New Dataset
new_bike_sharing_df = bike_sharing_df[bike_sharing_df["casual"] < upperBound]
new_bike_sharing_df.shape
```

- Melakukan proses capping pada dataset yang baru, sehingga dataset yang baru tidak memiliki outlier pada kolom "casual".

```python
# Capping the Dataset
new_bike_sharing_df_cap = bike_sharing_df.copy()
new_bike_sharing_df_cap["casual"] = np.where(
    new_bike_sharing_df_cap["casual"] > upperBound,
    upperBound,
    np.where(
        new_bike_sharing_df_cap["casual"] < lowerBound,
        lowerBound,
        new_bike_sharing_df_cap["casual"]
    )
)
```

**Insight:**
- Tidak terdapat format date and time yang tidak sesuai.
- Terdapat outlier pada kolom "casual", sehingga outlier pada kolom tersebut perlu di hapus.

## Clustering Analysis
#### Pertanyaan 4 : Bagaimana behaviour dari user rental sepeda terhadap temperature dan kecepatan angin yang berbeda-beda?

### KMeans Clustering for Total User with Respect to Windspeed
- Melakukan analisis lanjutan dengan menggunakan metode KMeans Clustering. Dimana variabel yang akan dianalisis adalah "cnt" dan "windspeed". Analisis ini diperlukan guna untuk mengetahui customer behaviour dari user peminjaman sepeda terhadap kecepatan angin yang berbeda-beda.

- Import library sklearn untuk menggunakan algoritma KMeans

```python
# Import Library
from sklearn.cluster import KMeans
```

- Melihat scatter plot dari total user terhadap kecepatan angin.

```python
# Scatter Plot for the Data
plt.scatter(new_bike_sharing_df_cap["windspeed"], new_bike_sharing_df_cap["cnt"])
plt.title("Correlation Between Total User with Respect to Windspeed")
plt.xlabel("Windspeed")
plt.ylabel("Total User")
plt.show()
```

- Membuat algoritma Kmeans dengan memanggil class KMeans dari scikit learn.

```python
# Define the KMeans
km = KMeans(n_clusters=4, random_state=0)
```

- Membuat salinan untuk dataset baru yang akan digunakan untuk keperluan Clustering

```python
# Define Dataset for Clustering
bike_sharing_clusters = new_bike_sharing_df_cap[["cnt", "windspeed"]].copy()
```

- Melakukan proses scaling data yang dibutuhkan agar fitur memiliki range nilai yang sama.

```python
# Scaling the Data
bike_sharing_clusters_scaled = (bike_sharing_clusters - bike_sharing_clusters.mean()) / bike_sharing_clusters.std()
```

- Memasukkan cluster ke dataset yang dikhususkan untuk analisis cluster.

```python
# Append the Cluster into the Dataset
clusters = km.fit(bike_sharing_clusters_scaled)
bike_sharing_clusters["clusters"] = clusters.labels_
```

- Melakukan plot untuk melihat persebaran dari cluster.

```python
# Plot the Clusters
plt.figure(figsize=(10,6))
plt.scatter(bike_sharing_clusters["windspeed"], bike_sharing_clusters["cnt"], c=bike_sharing_clusters["clusters"], cmap="viridis")
plt.title("Clusters of Total Users with Respect of Windspeed")
plt.xlabel("Windspeed")
plt.ylabel("Total User")
plt.show()
```

- Memvalidasi penggunaan nilai K yang optimal dengan menggunakan Elbow Method

```python
# Elbow Method Algorithm
wcss = []
for k in range(1,12):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(bike_sharing_clusters_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(11,6))
plt.plot(range(11), wcss)
plt.title("Elbow Method")
plt.xlabel("Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

**Insight**
- Dari grafik elbow method, terlihat bahwa titik penurunan inertia mulai melambat pada k = 3 dan k = 4. Oleh karena itu, nilai K yang optimal untuk clustering berada di 3 dan 4.
- Data terbagi menjadi empat bagian yaitu bagian dengan warna kuning, hitam, biru dan hijau.
- Cluster kuning merupakan representasi user yang lebih sedikit pada kecepatan angin yang lebih tinggi. User yang ada di cluster ini berkurang saat kecepatan angin lebih tinggi.
- Cluster hitam dan biru merepresentasikan bahwa pada kecepatan angin yang lebih rendah terjadi peningkatan user yang menggunakan rental sepeda. Cluster ini menunjukkan bahwa kecepatan angin yang sedang cukup nyaman untuk bersepeda.
- Cluster hijau merepresentasikan bahwa pada kecepatan angin yang sangat rendah terjadi lonjakan user yang lebih tinggi. Hal ini menunjukkan bahwa kecepatan angin yang sangat rendah sangat cocok untuk bersepeda.
